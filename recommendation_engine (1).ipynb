```python
from sklearn.neighbors import NearestNeighbors

# Load and preprocess the data
df = pd.read_csv('user_Profiles.csv')


X = df['Nutrient']
Y = df['Veg_Non']

# Define the similarity metric
def cosine_similarity(X, Y):
    return np.dot(X, Y) / (np.linalg.norm(X) * np.linalg.norm(Y))


# Fit the KNN model
k = 5
model = NearestNeighbors(n_neighbors=k, algorithm='brute', metric=cosine_similarity)
model.fit(X)

# Make recommendations
test_user_id = 80
test_user_ratings = df[df['User_80'] == test_user_id].drop(['User_80'], axis=1)
```
