```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('recent_activity.csv')
```

```python
# Univariate analysis - Histogram
plt.hist(df['User_Id'], bins=10)
plt.xlabel('User Id')
plt.ylabel('purchased')
plt.title('Recent activity')
plt.show()
```

```python
# Univariate analysis - Boxplot
sns.boxplot(x=df['Rated'])
plt.xlabel('Purchased')
plt.title('Purchased')
plt.show()
```

```python
# Bivariate analysis - Scatter plot
sns.scatterplot(x=df['Purchased'], y=df['Rated'])
plt.xlabel('Purchased')
plt.ylabel('Rated')
plt.title('Scatter Plot of Purchased vs Rated')
plt.show()
```

```python
# Bivariate analysis - Heatmap
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
